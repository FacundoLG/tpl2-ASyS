```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ejercicio 4: Sistemas LTI en tiempo discreto y continuo
### 4.1

```python
h = np.array([1,2,2,1])
n = np.arange(-1,3,1)
plt.grid(True)
plt.stem(n,h)
```

Ahora se descompone h dos funciones tales que: **h[n] = h1[n] + h2[n]**

```python
h1 = np.array([1,2,2,0])
h2 = np.array([0,0,0,1])
fig, axs = plt.subplots(2,1)
fig.tight_layout()
n = np.arange(-1,3)
axs[0].set_title("h1[n]")
axs[0].stem(n,h1)
axs[1].set_title("h2[n]")
axs[1].set_ylim(0,2)
axs[1].stem(n,h2)
```

```python
y = np.convolve(h1,h)
y1= np.convolve(h1,h1)
y2= np.convolve(h1,h2)

fig, axs = plt.subplots(4,1)
fig.tight_layout()
n = np.arange(-2,5)
axs[0].set_title("y1[n]")
axs[0].stem(n,y1)
axs[1].set_title("y2[n]")
axs[1].stem(n,y2)
axs[2].set_title("y1[n]+y2[n]")
axs[2].stem(n,y1+y2)
axs[3].set_title("y[n]")
axs[3].stem(n,y)
```

```python
w =y1
y = np.convolve(w,h2)
yp = np.convolve(h1,y2)

n = np.arange(-2,5)
n2 = np.arange(-3,7)

fig, axs = plt.subplots(3,1)
fig.tight_layout()

axs[0].set_title("w[n]")
axs[0].stem(n,w)

axs[1].set_title("y[n]")
axs[1].stem(n2,y)
axs[2].set_title("y[n]")
axs[2].stem(n2,yp)
```

Las salidas son iguales, por lo que las propiedades se cumplen
