## Ejercicio 2: Transformaciones de la variable independiente

### 2.1

```python
def u(t):
    return np.where(t>=0,1,0)

def x(t):
    return u(t+1)-u(t-2)+u(t)-u(t-1) #Expresion

t = np.linspace(-10,10,1000)
plt.grid(True)
plt.plot(t,x(t))
```

La expresion es: **u(t+1)-u(t-2)+u(t)-u(t-1)**

### 2.2 Realice una funcion capaz de determinar si una funcion matematica es par o impar

```python
def parimpar(func,dist=100,res=10000):
    """
    Parametros
    func - Funcion a evaluar
    dist - Distancia al origen, para crear un intervalo [-dist,dist]
    res  - Resolucion

    Retorna
    -1 - Si no es par ni impar
     2 - Si es par
     1 - Si es impar
    """
    def clasificador(x,i):
        if x(i) == -x(-i):
            return 1 
        elif x(i) == x(-i):
            return 2
        else:
            return -1
            

    t = np.linspace(0,dist,res)
    
    resultado_inicial = clasificador(func,t[1]) # Resultado de la primera clasificacion
    if resultado_inicial == -1: 
        return resultado_inicial

    for i in t:
        if(func(i) == 0 and func(-i) == 0):
            continue
        if clasificador(func,i) != resultado_inicial:
            return -1
         
    return resultado_inicial


def test(t):
    return u(t) - u(-t)

t = np.linspace(-20,20,1000)

plt.plot(t,test(t))
parimpar(test)
```
