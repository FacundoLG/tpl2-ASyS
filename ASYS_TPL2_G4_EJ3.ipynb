```python
import numpy as np
import matplotlib.pyplot as plt
```

## Ejercicio 3: Intergral y Suma de Convolucion
### 3.1.a

```python
def u(t):
    return np.where(t >=0,1,0)

def x(t):
    return u(t) -u(t-3)

def n(t):
    return u(t) -u(t-3)

t = np.linspace(-1,4,1000)

fig, axs = plt.subplots(2,1)
axs[0].set_title("x(t)")
axs[0].plot(t,x(t))
axs[1].set_title("n(t)")
axs[1].plot(t,n(t))
```

```python
t = np.linspace(0,3,1000)
sig1 = x(t)
sig2 = n(t)
conv = np.convolve(sig1, sig2) * (t[1] - t[0])

t_c = np.linspace(0,6,len(conv))
plt.plot(t_c,conv)
```

### 3.1.b

```python
t = np.linspace(-1,5,1000)
def x(t):
    return u(t-1)-u(t-4)

fig, axs = plt.subplots(2,1)
fig.tight_layout()
axs[0].set_title("x(t)")
axs[0].plot(t,x(t))
axs[1].set_title("n(t)")
axs[1].plot(t,n(t))
```

```python
t = np.linspace(0,4,1000)
sig1=x(t)
sig2=n(t)
conv = np.convolve(sig1,sig2,mode="full") * (t[1] - t[0])
c_t = np.linspace(0,8,len(conv))
plt.plot(c_t,conv)
```

### 3.1.c

```python
def x(t):
    return u(t) - u(t-5)

def n(t):
    return u(t) - u(t-2)

t = np.linspace(-1,5,1000)
fig, axs = plt.subplots(2,1)
fig.tight_layout()

axs[0].set_title("x(t)")
axs[0].plot(t,x(t))
axs[1].set_title("n(t)")
axs[1].plot(t,n(t))
```

```python
t = np.linspace(0,5,1000)
sig1=x(t)
sig2=n(t)
conv = np.convolve(sig1,sig2,mode="full") * (t[1] - t[0])
c_t = np.linspace(0,10,len(conv))
plt.grid(True)
plt.xticks(np.arange(0,10,1))
plt.plot(c_t,conv)
```

### 3.1.d

```python
def x(t):
    return u(t) - u(t-5)

def n(t):
    return 2*u(t) - 2*u(t-3)

t = np.linspace(-1,5,1000)
fig, axs = plt.subplots(2,1)
fig.tight_layout()

axs[0].set_title("x(t)")
axs[0].plot(t,x(t))
axs[1].set_title("n(t)")
axs[1].plot(t,n(t))
```

```python
t = np.linspace(0,5,1000)
sig1=x(t)
sig2=n(t)
conv = np.convolve(sig1,sig2,mode="full") * (t[1] - t[0])
c_t = np.linspace(0,10,len(conv))
plt.grid(True)
plt.xticks(np.arange(0,10,1))
plt.plot(c_t,conv)
```
